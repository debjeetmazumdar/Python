```python
def player1_option():
    
    #placeholders
    
    player1_option=True
    
    
    
    # player1 input validation
    
    while player1_option==True:    
        
        player1_option=input("Player1,select your spot from range 1-9")
        
        if player1_option.isdigit()==False:
            print("Invalid option!")            
        
        elif int(player1_option) not in range(1,10):            
            print("Selection out of range, select from range 1-9")
            
        else:
            pass
        

    #Return player selected options
    return int(player1_option)


    
```

```python

        
def player2_option():
    
    #placeholders
    
    player2_option=True
    
    
    # player2 input validation
    
    while player2_option==True:    
        
        player2_option=input("Player2,select your spot from range 1-9")
        
        if player2_option.isdigit()==False:
            print("Invalid option!")            
        
        elif int(player2_option) not in range(1,10):            
            print("Selection out of range, select from range 1-9")
            
        else:
            pass
            
    
      
    #Return player selected options
    return int(player2_option)
```

```python
def update_board_for_player1(board,option):
    
    board[option-1]='X'
    
    return board
    
```

```python
def update_board_for_player2(board,option):
    
    board[option-1]='O'
    
    return board
```

```python
def results(board):
    
    for x in [0,3,6]:
        if board[x]==board[x+1]==board[x+2]=='X':
            print("Player1 won!")
            return True
        elif board[x]==board[x+1]==board[x+2]=='O':
            print("Player2 won!")
            return True
        else:
            pass
    
    for x in [0,1,2]:
        if board[x]==board[x+3]==board[x+6]=='X':
            print("Player1 won!")
            return True
        elif board[x]==board[x+1]==board[x+2]=='O':
            print("Player2 won!")
            return True
        else:
            pass
        
        
    if board[2]==board[4]==board[6]=='X':
        print("Player1 won!")
        return True
    elif board[2]==board[4]==board[6]=='O':
        print("Player2 won!")
        return True
            
    
    if board[0]==board[4]==board[8]=='X':
        print("Player1 won!")
        return True
    elif board[0]==board[4]==board[8]=='O':
        print("Player2 won!")
        return True
        
        
    return False    
```

```python
def continuegame():
    
    decision=True   
    
    
    while decision:
        
        decision= input("Do you want to play again- Y or N")
        
        if decision.isdigit():
            print('Invalid input')
            
        elif (decision !='Y' and decision !='N'):
            print('Invalid input')
        
        elif decision =='Y':
            return True
            
        elif decision =='N':
            return False
        else:
            pass
        
            
        
        
    
    
```

```python
def game():
    

    
    board=display_board([' ',' ',' ',' ',' ',' ',' ',' ',' '])
    
    res=False
    
    bool=True
    
    while res==False:
        
        
        if res==False:
        
            option1=player1_option()
    
            updated_board=update_board_for_player1(board,option1)
        
            display_board(updated_board)
            
            res=results(updated_board)
            
            #print(res)
        
        
        if res==False:
                
            option2=player2_option()
    
            updated_board=update_board_for_player2(updated_board,option2)
        
            display_board(updated_board)
            
            res=results(updated_board)
                
                
    if res==True:
        
        bool=continuegame()
        
    if bool==True:
        
        game()
    else:
        print('Thanks')
    
    
    
    
```
